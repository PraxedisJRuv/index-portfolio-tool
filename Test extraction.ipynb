```python
#Try using Stooq
import pandas as pd
from datetime import datetime, timedelta

tickers = ["AAPL.US"]
end=datetime.today()
start=end-timedelta(days=2*365)
flag=True
for ticker in tickers:
    url=f"https://stooq.com/q/d/l/?s={ticker}&d1={start:%Y%m%d}&d2={end:%Y%m%d}&i=d"
    df_temp=(pd.read_csv(url,parse_dates=["Date"])
        .set_index("Date")
        .sort_index())
    if flag:
        data=df_temp
        flag =False
    else:
        data=pd.merge(data, df_temp, left_on=data.columns[0], right_on=df_temp.columns[0], how='outer')

    data
```

```python
start_date = '2015-01-01'
end_date = '2024-12-31'
# we could also request by period in an interval
#period=1, interval="1m"

#This if we want to save it in csv (I'm willing to change this to a MySQL thing)
for nombre, ticker in companies.items():
    try:
        print(f"Descargando datos de: {nombre} ({ticker})")
        data = yf.download(ticker, start=start_date, end=end_date)["Close"]
        #data= yf.download(tickers, period="1d", interval="1m")["Close"]
        if not data.empty:
            data.to_csv(f"{ticker}_{start_date}_to_{end_date}.csv")
        else:
            print(f"No se encontraron datos para {nombre} ({ticker})")
    except Exception as e:
        print(f"Error descargando {nombre} ({ticker}): {e}")

#This if we just want the close and nothing more
data = yf.download(tickers, period="1d", interval="1m")["Close"]
```
